```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris=load_iris(return_X_y=True, as_frame=True)
```

```python
type(iris)
```

```python
X=iris[0]
y=iris[1]
print(type(X), type(y))
```

```python
X.head()
```

```python
y.head()
```

```python
#define the data and target variables to be used in the model
X_train=X.iloc[:, 2:] #choose petal length and width
X_train.head()
```

```python
y_train=iris[1]
y_train.head()
```

```python
tree_clf=DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X_train, y_train)
```

```python
#visulalize the tree
iris_a=load_iris()
from sklearn import tree
from matplotlib import pyplot as plt
fig=plt.figure(figsize=(25,20))
tree_fig=tree.plot_tree(tree_clf, 
                   feature_names=iris_a.feature_names[2:],  
                   class_names=iris_a.target_names,
                   filled=True)
```

```python
#Make some predictions using class probabilities
tree_clf.predict_proba([[5, 1.5]]) #for petal length=5 and petal width=1.5
```
